**Importing Required Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**Create a Dataframe with Dependent Variable(x) and independent variable y.**

```python
x=np.array([95,85,80,70,60])
y=np.array([85,95,70,65,70])
```

**Create Linear Regression Model using Polyfit Function:**


```python
model= np.polyfit(x, y, 1)
```

**Observe the coefficients of the model.**


```python
model
```

**Predict the Y value for X and observe the output.**


```python
predict = np.poly1d(model)
```

```python
predict(65)
```

**Predict the y_pred for all values of x.**

```python
y_pred= predict(x)
```

```python
y_pred
```

**Evaluate the performance of Model (R-Suare)**

```python
from sklearn.metrics import r2_score
r2_score(y, y_pred)
```

**Plotting the linear regression model**



```python
y_line = model[1] + model[0]* x
plt.plot(x, y_line, c = 'r')
plt.scatter(x, y_pred)
plt.scatter(x,y,c='r')
```

**Filling null values**

```python
data['CRIM'] = data['CRIM'].fillna(data['CRIM'].mean())
data['ZN'] = data['ZN'].fillna(data['ZN'].mean())
data['INDUS'] = data['INDUS'].fillna(data['INDUS'].mean())
data['CHAS'] = data['CHAS'].fillna(data['CHAS'].mean())
data['AGE'] = data['AGE'].fillna(data['AGE'].mean())
data['LSTAT'] = data['LSTAT'].fillna(data['LSTAT'].mean())
```

```python
data.isnull().sum()
```

**Split dependent variable and independent variables**


```python
x = data.drop(['MEDV'], axis = 1)
x
```

```python
y = data['MEDV']
y
```

**Splitting data to training and testing dataset.**

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size =0.2,random_state = 0)
```

**Use linear regression( Train the Machine ) to Create Model**

```python
import sklearn
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
model=lm.fit(xtrain, ytrain)
```

```python
model
```

**Predict the y_pred for all values of train_x and test_x**

```python
ytrain_pred = lm.predict(xtrain)
ytrain_pred
```

```python
ytest_pred = lm.predict(xtest)
ytest_pred
```

```python
df=pd.DataFrame(ytrain_pred,ytrain)
df
```

**Calculate Mean Square Paper for train_y and test_y**


```python
from sklearn.metrics import mean_squared_error, r2_score
mse_test = mean_squared_error(ytest, ytest_pred)
print(mse_test)
mse_train = mean_squared_error(ytrain_pred,ytrain)
print(mse_train)
```

```python
rmse_test=np.sqrt(mse_test)
print(rmse_test)
```

```python
rmse_train=np.sqrt(mse_train)
print(rmse_train)
```

**Plotting the linear regression model**

```python
plt.scatter(ytrain ,ytrain_pred,c='blue',marker='o',label='Training data')
plt.scatter(ytest,ytest_pred ,c='lightgreen',marker='s',label='Test data')
plt.xlabel('True values')
plt.ylabel('Predicted')
plt.title("True value vs Predicted value")
plt.legend(loc= 'upper left')
plt.plot()
plt.show()
```

```python
r2_score_test=r2_score(ytest,ytest_pred)
print(r2_score_test)
```
